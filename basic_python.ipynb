```python
#simple interest
Rate = int(input("enter the rate "))
principal=int(input("enter the principle amount"))
time=int(input("enter the time  in month "))
month_name =input("enter name of month")
time_in_days=int(input("enter the time  in days "))
days=int(input("enter the days "))

if month_name=="januray" or month_name=="march" or month_name=="may" or  month_name=="july" or  month_name=="august" or month_name=="october" or month_name=="december":
  days_in_month=31
elif month_name=="april" or month_name=="june" or month_name=="september" or month_name=="november" :
  days_in_month=30
elif month_name=="february":
  if days%400==0 or (days%4==0 and days%100 !=0):
       days_in_month=29

  else:
    days_in_month=28


#convert days inot the month
time_in_month=time_in_days/days_in_month

simple_interst1=(Rate * principal * time) / 100
simple_interst2=(Rate * principal * time_in_month) / 100

SI=simple_interst1+simple_interst2
print("Total simple interest is :",SI)
```
